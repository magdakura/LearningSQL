## STRING MANIPULATION

```python
Q("""
SELECT 
    starttime, 
    substr(starttime, 1,10) as date,
    substr(starttime, 1,4) as year, 
    substr(starttime, 6,2) as month, 
    substr(starttime, 9,2) as day, 
    substr(starttime, 11,10) as time
FROM tripdata
LIMIT 3
""")
```

```python
Q("""
SELECT 
    bikeid, 
    substr(starttime, 6,2) AS month, 
    sum(tripduration)/3600 AS total 
FROM 
    tripdata 
GROUP BY 
    bikeid,
    month 
ORDER BY total""")
```

```python
Q("""
SELECT 
    bikeid,
    sum(case when month = '06' then total end) as jun,
    sum(case when month = '07' then total end) as jul,
    sum(case when month = '08' then total end) as aug,
    sum(case when month = '09' then total end) as sept
from
(select
    bikeid, 
    substr(starttime, 6,2) AS month, 
    sum(tripduration)/3600 AS total 
FROM 
    tripdata 
GROUP BY 
    bikeid,
    month 
ORDER BY total)
GROUP BY bikeid
    """)
```

## pivot table on gender

```python
Q("""
SELECT
    start_station_name,
    sum(case when gender=1 then 1 else 0 end) as male,
    sum(case when gender=2 then 1 else 0 end) as female
FROM tripdata
GROUP BY 1
""")
```

```python
Q("""
SELECT
    start_station_name,
    100.0 * male / (male + female) as pct_male
FROM
(
SELECT
    start_station_name,
    sum(case when gender=1 then 1 else 0 end) as male,
    sum(case when gender=2 then 1 else 0 end) as female
FROM tripdata
GROUP BY 1
)
""")
```

```python
Q("""
SELECT
    start_station_name,
    round( (2017 - birth_year) / 5) * 5 as age,
    substr(starttime, 11, 2) as hour,
    sum(case when gender=1 then 1 else 0 end) as male,
    sum(case when gender=2 then 1 else 0 end) as female
FROM tripdata
GROUP BY 1, 2, 3
""")
```
