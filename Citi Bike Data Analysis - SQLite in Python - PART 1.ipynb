```python
import pandas
```

```python
import sqlite3
```

```python
pandas.read_csv("https://s3.amazonaws.com/tripdata/JC-201709-citibike-tripdata.csv.zip", nrows=10)
```

```python
db = sqlite3.connect('citibike.db')
```

```python
url = "https://s3.amazonaws.com/tripdata/JC-20170%s-citibike-tripdata.csv.zip"
for i in (5,6,7,8,9):
    print (url % i)
    chunks = pandas.read_csv(url %i, chunksize=100_000)
    for chunk in chunks:
        chunk.columns = [ column.replace(' ','_') for column in chunk.columns]
        chunk.to_sql('tripdata', db, if_exists='append')
```

```python
pandas.read_sql_query('SELECT * FROM tripdata LIMIT 10', db)
```

```python
pandas.read_sql_query('SELECT avg(tripduration) FROM tripdata', db)
```

```python
pandas.read_sql_query('SELECT count(*) FROM tripdata GROUP BY gender', db)
```

```python
pandas.read_sql_query('SELECT gender, usertype,count(*) FROM tripdata GROUP BY gender, usertype', db)
```

```python
pandas.read_sql_query('SELECT gender, avg(tripduration) FROM tripdata GROUP BY gender',db)
```

```python
pandas.read_sql_query('SELECT (2017- birth_year) AS age, count(*) FROM tripdata GROUP BY age HAVING age>1', db)
```
